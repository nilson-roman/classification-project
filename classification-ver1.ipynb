```python
X = analysis_tmdb.drop(['vote_average', 'vote_average_enc'], axis=1)
X = pd.get_dummies(X, drop_first=True)
y = analysis_tmdb.vote_average_enc
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, test_size=0.2)
```

```python
model = LogisticRegressionCV()
model.fit(X_train, y_train)
```

```python
lc = LearningCurve(model,scoring='precision', random_state=42)
lc.fit(X,y)
lc.show()
```

```python
y_train.mean()
```

```python
y_pred_train = model.predict(X_train)
y_pred_test = model.predict(X_test)
```

```python
print(classification_report(y_train, y_pred_train))
```

```python
print(classification_report(y_test, y_pred_test))
```

```python
confusion_matrix(y_train, y_pred_train)
```

# Shap   (lgbm model)

```python
model = LGBMClassifier()
model.fit(X_train,y_train)
```

```python
single_row = X_test.iloc[[1],:]
```

```python
explainer = shap.TreeExplainer(model=model, data=X_train)
```

```python
values = explainer.shap_values(single_row)
```

```python
shap.force_plot(base_value=explainer.expected_value, shap_values=values,features = single_row)
```
